```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('covid_data.csv')

# Show first few rows
print(data.head())

# Summary statistics
print(data.describe())

# Group by country and sum confirmed cases
country_cases = data.groupby('Country')['Confirmed'].sum().sort_values(ascending=False)

# Plot top 5 countries
country_cases.head(5).plot(kind='bar', figsize=(8,5))
plt.title('Top 5 Countries with Most COVID-19 Cases')
plt.xlabel('Country')
plt.ylabel('Total Confirmed Cases')
plt.show()
```
